```
import numpy  as np
from scipy.optimize import minimize
from scipy import optimize
```

```
def RDTR_calc(x): 
    y =  5.66-0.21*x[0]**2+0.17*x[1]-0.20*x[2]**2+0.21*x[3]-0.47*x[3]**2+0.24*x[0]*x[1]+0.43*x[0]*x[3]-0.40*x[1]*x[3]
    return y
def RDTE_calc(x):
    y =  3.87+0.38*x[0]-0.39*x[0]**2-0.31*x[1]**2-0.22*x[2]**2+0.16*x[3]-0.45*x[3]**2+0.53*x[1]*x[3]
    return y
def RDCOT_calc(x):
    y=6.10-0.24*x[0]**2-0.49*x[1]**2-0.24*x[2]**2+0.53*x[3]-0.46*x[3]**2+0.85*x[0]*x[1]-0.36*x[0]*x[2]+0.21*x[0]*x[3]+0.73*x[1]*x[2]+0.36*x[1]*x[3]+0.50*x[2]*x[3]  
    return y
def RDEXP_calc(x):
    y=6.14+0.17*x[0]-0.16*x[0]**2+0.05*x[1]-0.24*x[1]**2+0.20*x[2]-0.04*x[2]**2-0.07*x[3]**2+0.17*x[1]*x[3]+0.11*x[2]*x[3]
    return y    
def RDRED_calc(x):
    y=6.02+0.23*x[0]-0.08*x[1]**2+0.31*x[3]-0.27*x[3]**2+0.15*x[0]*x[1]-0.10*x[0]*x[2]+0.31*x[0]*x[3]+0.37*x[1]*x[3]+0.15*x[2]*x[3]
    return y
def Energy_calc(x):
    y=124.62+37.28*x[0]-3.08*x[0]**2+12.86*x[1]-4.86*x[1]**2-4.57*x[2]
    return y            
def Desirability(y,U,T,L):
    if y<L:
        d=0
    if (y>=L)&(y<=T):
        d=(y-L)/(T-L)
    if (y>=T)&(y<+U):
        d=(U-y)/(U-T)
    if (y>U):
        d=0
    return d
```

```
def function(x):
    RDTR=RDTR_calc(x)
    RDTE=RDTE_calc(x)
    RDCOT=RDCOT_calc(x)
    RDEXP=RDEXP_calc(x)
    RDRED=RDRED_calc(x)
    print("===========")
    print(RDTR)
    print(RDTE)
    print(RDCOT)
    print(RDEXP)
    print(RDRED)

    U=7
    T=6
    L=3
    d1=Desirability(RDTR,U,T,L)
    d2=Desirability(RDTE,U,T,L)
    d3=Desirability(RDCOT,U,T,L)
    d4=Desirability(RDEXP,U,T,L)
    d5=Desirability(RDRED,U,T,L)

    Energy=Energy_calc(x)
    U1=300
    T1=120
    L1=0
    d6=Desirability(Energy,U1,T1,L1)

    fun = (-1)*(d1*d2*d3*d4*d5)**(1/5)
    #fun = (-1)*(d2)**(1/5)
    print(fun)
    return fun
```

```
bounds=[(-2,2),(-2,2),(-2,2),(-2,2)]
Condicoes = optimize.differential_evolution(function,bounds,init="latinhypercube",maxiter=1000, popsize=80,recombination=0.5)
x_opt = Condicoes.x
x_opt
```
